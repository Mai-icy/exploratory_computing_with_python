### 除法
除法也可以使用

```python
print('1/3 gives', 1 / 3)
```

（针对 Python 2 用户的说明（你真的应该切换到 Python 3！）：在 Python 2 中，1/3 的结果为零，因为两个整数的除法在 Python 2 中返回一个整数。）上面的打印语句在连续显示 16 个 3 的值时看起来非常丑陋。更好且更易读的方法是使用称为 f-string 的格式来同时打印文本和变量的值。f-string 允许你通过用大括号 {} 包围变量的值，将其插入到文本的任何位置。整个文本字符串需要用引号括起来，并且前面需要加上字母 f，就像下面。

```python
a = 1
b = 3
c = a / b
print(f'{a} divided by {b} gives {c}')
```

大括号中的完整语法是 {variable:width.precision}。当未指定 输出总长 和 保留小数位 时，Python 会使用所有数字并为你计算宽度。如果你想要一个小数点后有 3 位的浮点数，可以指定数字 3，后面跟上字母 f 表示浮点数（你仍然可以不指定宽度，让 Python 自动计算）。如果你更喜欢指数（科学）表示法，可以用 e 替换 f。 # 后面的文本是代码中的注释。在 # 后的任何文本都会被 Python 忽略。

```python
print(f'{a} divided by {b} gives {c:.3f}') # 保留位小数
print(f'{a} divided by {b} gives {c:10.3f}') # 长度为10 保留三位小数
print(f'{a} divided by {b} gives {c:.3e}') # 三位小数用科学计数法
```

### 风格

正如上面提到的，良好的编码风格非常重要。它使代码更易于阅读，从而更容易发现错误和漏洞。例如，考虑下面的代码，它重新创建了我们之前绘制的图形（使用更宽的线条），但现在没有插入额外的空格。（很挤很难看吧）

```python
a=1
b=1
c=-6
x=np.linspace(-4,4,100)
y=a*x**2+b*x+c#Compute y for all x values
plt.plot(x,y,linewidth=3)
```

在前面的代码单元中，代码难以阅读。良好的风格至少包括以下几点：

- 每个数学符号（=, +, -, *, /）周围都有空格，但 ** 周围不需要空格
- 函数参数之间有空格
- 关键字参数的等号两边没有空格（因此 linewidth=3 是正确的）
- 每个逗号后有一个空格
- 每个 # 后有一个空格
- 当 # 紧跟在 Python 语句后面时，前面有两个空格
- 函数名和参数列表之间没有空格。因此 plt.plot(x, y) 是良好的风格，而 plt.plot (x, y) 就不符合良好的风格。

这些规则是官方 Python 风格指南 PEP8 的（非常小的一部分）。当这些规则被应用时，代码可读性会大大提高，如下所示：

```python
a = 1
b = 1
c = -6
x = np.linspace(-4, 4, 100)
y = a * x**2 + b * x + c  # Compute y for all x values
plt.plot(x, y, linewidth=3);
```

在所有其他练习和未来的 Notebook 中使用正确的风格。

### <a name="ex4"></a> 练习 4，子图和花哨的标记

```python
air_temp = np.loadtxt('holland_temperature.dat')
sea_temp = np.loadtxt('holland_seawater.dat')

months = ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']

plt.figure(figsize=(10, 8))

plt.subplot(2, 1, 1)
plt.plot(air_temp, label='Air Temperature', color='blue', linewidth=2)
plt.plot(sea_temp, label='Sea Temperature', color='red', linewidth=2)
plt.xticks(ticks=np.arange(12), labels=months)
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.title('Average Monthly Air and Sea Temperature in Holland')
plt.legend()
plt.grid()

plt.subplot(2, 1, 2) 
temperature_difference = air_temp - sea_temp
plt.plot(temperature_difference, label='Temperature Difference', color='green', linewidth=2)
plt.xticks(ticks=np.arange(12), labels=months) 
plt.xlabel('Month')
plt.ylabel('Temperature Difference (°C)')
plt.title('Temperature Difference between Air and Sea')
plt.legend()
plt.grid()

plt.tight_layout()
plt.show()
```

```python
countries = ['USA', 'CHN', 'GBR', 'RUS', 'KOR', 'GER', 'FRA', 'ITA', 'HUN', 'AUS', 'OTHER']
gold_medals = [46, 38, 29, 24, 13, 11, 11, 8, 8, 7, 107]

colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0',
          '#ffb3e6','#ff6666','#c2f0c2','#ffccf2','#c2c2c2','#c0c0c0']

plt.figure(figsize=(10, 8))
plt.pie(gold_medals, labels=countries, autopct='%1.1f%%', startangle=140, colors=colors)
plt.axis('equal')

plt.title('Gold Medals Won by Country at the 2012 London Olympics')
plt.show()
```

```python
air = np.loadtxt('holland_temperature.dat') 
sea = np.loadtxt('holland_seawater.dat')
plt.fill_between(range(1, 13), air, color='b', alpha=0.4)
plt.fill_between(range(1, 13), sea, color='r', alpha=0.4)
plt.xticks(np.arange(1, 13), ['jan', 'feb', 'mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', ' oct', 'nov', 'dec'])
plt.xlabel('Month')
plt.ylabel('Temperature (Celcius)');
```
